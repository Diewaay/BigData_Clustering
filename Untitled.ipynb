```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('udemy_lang.csv')
```

```python
print(data.head())
```

```python
# Memilih kolom yang relevan untuk clustering
selected_columns = data[['rating', 'num_reviews', 'num_published_lectures']]
```

```python
# Menangani data yang hilang atau tidak valid
selected_columns = selected_columns.dropna()
```

```python
# Menstandarisasi data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_columns)
```

```python
# Mengurangi dimensi data menggunakan PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
# Visualisasi data setelah PCA
plt.scatter(pca_data[:, 0], pca_data[:, 1], s=50)
plt.title("Data setelah PCA")
plt.show()
```

```python
# Menggunakan K-means clustering
kmeans = KMeans(n_clusters=3, n_init=10)
cluster_labels = kmeans.fit_predict(pca_data)
```

```python
# Menambahkan label kluster ke dataset asli
data['cluster'] = cluster_labels
```

```python
# Menampilkan beberapa baris dari dataset dengan label kluster
print(data[['rating', 'num_reviews', 'num_published_lectures', 'cluster']].head())
```

```python
# Visualisasi kluster
plt.figure(figsize=(10, 7))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=cluster_labels, cmap='viridis', s=50)
plt.title("KMeans Clustering dengan PCA")
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Melihat distribusi course udemy dalam setiap cluster
cluster_distribution = data['cluster'].value_counts()
print(cluster_distribution)
```
